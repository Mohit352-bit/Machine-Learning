# Acne Detection Model

This model detects whether the choosen image of human face has acne or not. I used **Acne data** from the kaggle and for **Acne Free** I chose the images of actors as they take care and value their facial skin a lot and also use cosmetics to remove acne so their face will be perfect for acne free face data.I used CNN , transferlearning and since data was not enough I also used augmentaion.

### Transfer-Learning and Model building

```python
mod = tf.keras.applications.InceptionV3(
    include_top=True,
    weights="imagenet",
)
```

```python
ip = mod.layers[0].input
op = mod.layers[-2].output
op
```

```python
final_output=tf.keras.layers.Dense(128)(op)
final_output=tf.keras.layers.Activation('relu')(final_output)
final_output=tf.keras.layers.Dropout(0.2)(final_output)
final_output=tf.keras.layers.Dense(1,activation='sigmoid')(final_output)

final_output
```

```python
model=tf.keras.Model(inputs =ip,outputs=final_output)
model.summary()
```

```python
model.compile(loss = tf.keras.losses.BinaryCrossentropy(from_logits=True),
             optimizer= 'Adam',
             metrics=['accuracy'])
```

```python
train_gen = ImageDataGenerator(
    rescale = 1/255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2, 
    zoom_range=0.2,
    fill_mode='nearest'
)

train_data_gen = train_gen.flow_from_directory(train_data , 
                                              target_size = (299,299),
                                              batch_size=10,
                                              class_mode="binary")

val_gen = ImageDataGenerator(rescale = 1/255)
val_data_gen = val_gen.flow_from_directory(val_data ,
                                             target_size = (299,299),
                                             batch_size = 10,
                                             class_mode = "binary")
```

```python
categories = train_data_gen.class_indices.keys()
key_list = list(train_data_gen.class_indices.keys())
val_list = list(train_data_gen.class_indices.values())
print(key_list)
print(val_list)
```

```python
model_fiting = model.fit(train_data_gen ,steps_per_epoch=300,epochs=15,validation_data=val_data_gen, verbose=2 ,batch_size =10 )
```
