```python
# Principal Component Analysis is an unsupervised learning algorithm that is used for the dimensionality reduction in
# machine learning .It is a statistical process that converts the observations of correlated features into a set of 
# linearly uncorrelated features with the help of orthogonal transformation.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'] , columns = cancer['feature_names'])
df
```

```python
# to do principal component analysis we need to first standardise the data to a single scale.
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
c_data = scaler.transform(df)
```

```python
c = pd.DataFrame(c_data, columns = cancer['feature_names'])
```

```python
c
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
# here we are taking 2 as n_comp because we want to compress the data to 2 columns and get the most related data 
# on that two column so that the prediction of cancer is still near same
# It has converted most related data which causes cancer to group in 2 columns
pca = PCA(n_components=2)
```

```python
pca.fit(c_data)
```

```python
pca_d = pca.transform(c_data)
```

```python
plt.scatter(pca_d[:,0] , pca_d[:,1 ] ,c=cancer['target'])
```

## LogisticRegrr Using PCA

```python
X = pca_d
y = cancer['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
predict = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test , predict))
```
