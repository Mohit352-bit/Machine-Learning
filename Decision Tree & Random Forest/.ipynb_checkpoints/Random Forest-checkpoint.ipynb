```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
loan= pd.read_csv('loan_data.csv')
loan.head()
```

```python
# purpose column has 4 variation so we need to use get_dummies to make it readable by model

loan1 = pd.get_dummies(loan ,columns=['purpose'], drop_first=True)
loan1.dropna()
loan1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= loan1.drop('not.fully.paid' , axis=1)
y= loan[['not.fully.paid']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=120)
```

```python
rf.fit(X_train ,y_train)
```

```python
predict = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test , predict))
```
